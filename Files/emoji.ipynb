```python
import emoji

# emoji.demojize() - converts emoji to text
# emoji.version() - returns the version of the emoji

# print(joined_string)
first_num = int('0000', 16)
last_num = int('FFFF', 16)
get_range = range(first_num, last_num)
chars = [chr(c) for c in get_range]
unified_string = ''.join(chars)
# print(chars)
# print(unified_string)

only_emoji = []
for char in unified_string:
    if emoji.is_emoji(char):
        only_emoji.append(char)
print(''.join(only_emoji))

md_table = ''
md_table += '| Code | Name | Emoji |\n'
md_table += '| ---- | ---- | ----- |\n'

emoji_info = []
for char in only_emoji:
    char_props = {
        'code': ord(char),
        'name': emoji.demojize(char),
        'emoji': char
    }
    emoji_info.append(char_props)

for emoji_item in emoji_info:
    md_table += f'| {emoji_item["code"]:X} '
    md_table += f'| {emoji_item["name"]} '
    md_table += f'| {emoji_item["emoji"]} '
    md_table += '|\n'

# print(md_table)
```
